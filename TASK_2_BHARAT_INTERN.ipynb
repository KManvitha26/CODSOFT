```python
df['quality']=df['quality'].apply(lambda x:1 if x>=7 else 0)
```

```python
random_value=1000
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x=df[['fixed acidity', 'volatile acidity', 'sulphates', 'alcohol', 'density']]
y=df.quality
oversample=SMOTE()
x_ros,y_ros=oversample.fit_resample(x,y)

sns.countplot(x=y_ros)
plt.xticks([0,1],['bad wine', 'good wine'])
plt.title('Types of wine')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x_ros,y_ros, test_size=0.2, random_state=42)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm=MinMaxScaler()
norm_fit=norm.fit(x_train)
new_xtrain=norm_fit.transform(x_train)
new_xtest=norm_fit.transform(x_test)
print(new_xtrain)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error
```

```python
rfc=RandomForestClassifier()
fit_rfc=rfc.fit(new_xtrain,y_train)

score_rfc=rfc.score(new_xtest,y_test)
print('score of model is ', score_rfc)

y_predict = list(rfc.predict(x_test))

print('calculating the error')
mse_rfc=mean_squared_error(y_test,y_predict)
rmse_rfc=np.sqrt(mse_rfc)

print('mean squared error is ',mse_rfc)
print('root mean squared error is ',rmse_rfc)
print()
print(classification_report(y_predict,y_test))
```

```python
predicted_df = {'predicted_values': y_predict, 'original_values': y_test}
pd.DataFrame(predicted_df).head(20)
```
